##### **Problema 1**

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

plt.figure(figsize=(6, 4))

# Parámetros
media = 70
desv_std = 4
total_estudiantes = 500

limite_inferior = 68
limite_superior = 72

# Probabilidad entre 68 y 72
prob_68_72 = norm.cdf(limite_superior, media, desv_std) - norm.cdf(limite_inferior, media, desv_std)

# Número de estudiantes
estudiantes_68_72 = total_estudiantes * prob_68_72

print("Probabilidad de que un estudiante pese entre 68 y 72 kg:", prob_68_72)
print("Número de estudiantes que pesan entre 68 y 72 kg:", estudiantes_68_72)

# Gráfica
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= limite_inferior) & (x <= limite_superior), color='skyblue', alpha=0.5)

plt.title('Probabilidad de estudiantes entre 68 kg y 72 kg')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
