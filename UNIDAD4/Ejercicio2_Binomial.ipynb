##### **EJERCICIO 2 Binomial**

```python
from scipy.stats import binom

#Parámetros de la distribución binomial
n=10
p=0.10
x=2

#Calcular lla probabilidad de la variable aleatoria X=2
probabilidad_x_2 = binom.pmf(x, n, p)

print("La probabulidad de que la variable X sea igual a 2 es:  ", probabilidad_x_2)
```
