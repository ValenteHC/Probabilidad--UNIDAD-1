```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

plt.figure(figsize=(6, 4))

#Parámetros de la distibución  normal
media= 70 #Media
desv_std= 4 #Desviación estándar
total_estudiantes= 500
peso_limite=75
prob_menos_75kg=norm.cdf(peso_limite,media,desv_std)

estudiantes_menos_75kg=total_estudiantes*prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75kg:",prob_menos_75kg)
print("Número de estudiantes que pesan menos de 75kg:", estudiantes_menos_75kg)

x=np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y=norm.pdf(x,media,desv_std)

plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=( x <= peso_limite), color ='skyblue',alpha=0.5)
plt.title(r'Probabilidad de que un estudiante pese menos de 75kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

plt.figure(figsize=(6, 4))

#Parámetros de la distibución  normal
media= 70 #Media
desv_std= 4 #Desviación estándar
total_estudiantes= 500
peso_limite=79

#En esstas dos partes definimos los valores de "Probabilidad menor a 79kg" y "Estudiantes que pesen menos de 79kg".
prob_menos_79kg = norm.cdf(peso_limite,media,desv_std)
estudiantes_menos_79kg = total_estudiantes * prob_menos_79kg

#Aquí colocamos un mensaje antes de mosytrar la gráfica.
print("Probabilidad de que un estudiante pese menos de 79kg:",prob_menos_79kg)
print("Número de estudiantes que pesan menos de 79kg:", estudiantes_menos_79kg)

#Aquí construimos los puntos para formar la capana
x=np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y=norm.pdf(x,media,desv_std)

#Aquí le damos diseño a la gráfica, como colores, titulos, etc.
plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=( x >= peso_limite), color ='skyblue',alpha=0.5)
plt.title(r'Probabilidad de que un estudiante pese menos de 79kg.  $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()

```
