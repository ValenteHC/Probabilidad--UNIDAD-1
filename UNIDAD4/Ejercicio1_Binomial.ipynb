```python
from scipy.stats import binom

#Parámetros de la distribución binomial
probabilidad_exito=0.83
num_pacientes=16
X=16 #Num de exitos

#Calcular la probabilidad de que ningún paciente sufyra la enfermedad
probabilidad_ninguno_enfermo = binom.pmf(X, num_pacientes, probabilidad_exito)

print("La probabulidad de que ningún paciente sufra la enfermedad es: ", probabilidad_ninguno_enfermo)
```

```python
from scipy.stats import binom

#Parámetros de la distrubución binomial
pribabilidad_exito = 0.83
num_pacientes = 16
X=0

#Calcular la probabilidad de que todos los pacientes sufran la enfermedad
probabilidad_todos_enfermos = binom.pmf(X, num_pacientes, probabilidad_exito)
print("La probabulidad de que todos los pacientes sufran la enfermedad es: ", probabilidad_todos_enfermos)
```
