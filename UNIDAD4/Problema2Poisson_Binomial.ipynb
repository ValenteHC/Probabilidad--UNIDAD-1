##### **Problema 2 Binomial/Poisson**

```python
from scipy.stats import binom

 # Número total de ensayos
n_motores = 1000
 # Probabilidad de éxito (que sea defectuoso)
p_defecto = 0.001
print("Tipo de distribución: Binomial")
print("Justificación:")
print("- Se tienen n ensayos independientes (1000 motores)")
print("- Cada motor puede ser defectuoso o no (éxito/fracaso)")
print("- La probabilidad de defecto es constante (p = 0.001)\n")

print("Nota:")
print("También se puede aproximar con Poisson porque n es grande y p es pequeño.")
print("λ = n * p = 1000 * 0.001 = 1\n")
print(f"--- PARÁMETROS BINOMIALES ---")
print(f"n = {n_motores}")
print(f"p = {p_defecto}\n")
prob_0 = binom.pmf(0, n_motores, p_defecto)
prob_3 = binom.pmf(3, n_motores, p_defecto)
prob_4 = binom.pmf(4, n_motores, p_defecto)

#Mostrar resultados
print("--- RESULTADOS DEL PROBLEMA 2 (BINOMIAL) ---")
print(f"a) Probabilidad de que ninguno sea defectuoso: {prob_0:.4f}")
print(f"b) Probabilidad de que tres sean defectuosos:    {prob_3:.4f}")
print(f"c) Probabilidad de que cuatro sean defectuosos:   {prob_4:.4f}")

#Comparativa en porcentaje
print(f"\nResultados en porcentaje:")
print(f"a) {prob_0*100:.2f}%")
print(f"b) {prob_3*100:.2f}%")
print(f"c) {prob_4*100:.2f}%")
```
