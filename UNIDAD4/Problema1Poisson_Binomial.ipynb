##### **Problema 1 Binomial/Poisson**

```python
from scipy.stats import poisson
print("--- JUSTIFICACIÓN ---")
print("Tipo de distribución: Poisson")
print("Justificación: Se usa Poisson porque modela el número de eventos (ingresos)")
print("en un intervalo fijo de tiempo (un día), con una tasa promedio conocida (λ = 5),")
print("y suponiendo que los eventos ocurren de manera independiente.")
media_ingresos = 5
x_a = 2
prob_exactamente_2 = poisson.pmf(x_a, media_ingresos)
x_limite = 10
prob_exceda_10 = 1 - poisson.cdf(x_limite, media_ingresos)

print("RESULTADOS DEL PROBLEMA 1")
print(f"a) La probabilidad de que ingresen exactamente 2 personas es: {prob_exactamente_2:.4f}")
print(f"   (Es decir, un {prob_exactamente_2*100:.2f}%)")
print(f"b) La probabilidad de que x exceda de 10 es: {prob_exceda_10:.4f}")
print(f"   (Es decir, un {prob_exceda_10*100:.2f}%)")

if prob_exceda_10 < 0.05:
    print("\nExplicación: No es muy probable que exceda de 10 personas,")
    print("ya que la probabilidad es menor al 5%.")
else:
    print("\nExplicación: Existe una probabilidad considerable de que exceda de 10.")
```
