##### **PROGRAMA 4: Función de distribución acumuada**

```python
import math
import numpy as np
import matplotlib.pyplot as plt


x = [5,6,7,8,9,10]
p = [0.05,0.10,0.20,0.30,0.25,0.10]

#Calcular función acumulada
F= np.cumsum(p)

#Mostrar resultados
for xi, fi in zip(x,F):
  print(f"F({xi})={fi:.2f}")

#Graficar
plt.step(x,F,where='post')
plt.title("Función de distribución acumulada")
plt.xlabel("Calificación")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0,1.05)
plt.show()
```
