```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
px = np.array([0.05, 0.10, 0.25, 0.30, 0.20, 0.10])

esperanza = np.sum(x * px)
varianza = np.sum((x*2) * px) - esperanza*2
desviacion = np.sqrt(varianza)

print(f"E(X): {esperanza:.2f}")
print(f"Varianza: {varianza:.4f}")
print(f"Desviación Estándar: {desviacion:.4f}")

fx = np.cumsum(px)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.bar(x, px, color='orange', alpha=0.7, edgecolor='darkorange')
plt.title('Distribución de Ventas Semanales')
plt.xlabel('Unidades')
plt.ylabel('P(X)')

plt.subplot(1, 2, 2)
plt.step(np.append(x, 6), np.append(fx, 1.0), where='post', color='purple')
plt.title('Función Acumulada F(X)')
plt.xlabel('Unidades')
plt.ylabel('F(X)')
plt.grid(True, alpha=0.2)

plt.tight_layout()
plt.show()
```
