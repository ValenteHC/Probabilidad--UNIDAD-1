```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

datos = {
    'X': [0, 1, 2, 3, 4, 5],
    'P(X)': [0.10, 0.25, 0.30, 0.20, 0.10, 0.05]
}

df = pd.DataFrame(datos)

suma_prob = df['P(X)'].sum()
print(f"1. Verificación (Suma P(X)): {suma_prob}")

df['x_PX'] = df['X'] * df['P(X)']
valor_esperado = df['x_PX'].sum()

df['x2_PX'] = (df['X']**2) * df['P(X)']
varianza = df['x2_PX'].sum() - (valor_esperado**2)
desviacion_estandar = np.sqrt(varianza)

print(f"3. E(X) [Promedio]: {valor_esperado:.2f}")
print(f"   Varianza: {varianza:.4f}")
print(f"   Desviación Estándar: {desviacion_estandar:.4f}")

df['F(X)'] = df['P(X)'].cumsum()
print("\n4. Distribución de Probabilidad y Acumulada:")
print(df[['X', 'P(X)', 'F(X)']])

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(df['X'], df['P(X)'], color='skyblue', edgecolor='navy')
plt.title('Distribución de Probabilidad P(X)')
plt.xlabel('Células anormales')
plt.ylabel('Probabilidad')

plt.subplot(1, 2, 2)
x_step = np.append(df['X'].values, 6)
y_step = np.append(df['F(X)'].values, 1.0)
plt.step(x_step, y_step, where='post', color='red')
plt.title('Función Acumulada F(X)')
plt.xlabel('Células anormales')
plt.ylabel('F(X)')

plt.tight_layout()
plt.show()
```
