##### **PROGRAMA 5: Gráfica de distribución de probabilidad discreta**

```python
#-----------------------------------------------------
# Gráfica de una distribución de probabilidad discreta
# con media, varianza y desviasión estándar
#-----------------------------------------------------
import matplotlib.pyplot as plt
import numpy as np

# Datos: valores de la variable aleatoria y sus probabilidades
x = np.array([5,6,7,8,9,10])
p = np.array([0.05,0.10,0.20,0.30,0.25,0.10])

esperanza=np.sum(x*p)
esperanza_cuadrado=np.sum(x**2*p)
varianza=esperanza_cuadrado-esperanza**2
desviacion=np.sqrt(varianza)

#Mostrar los resultados en la consola
print(f"E(X) ={esperanza:.2f}")
print(f"E(X^2) ={esperanza_cuadrado:.2f}")
print(f"Varianza ={varianza:.2f}")
print(f"Desviación estándar ={desviacion:.2f}")

#Crear una gráfica tipo 'stem' (tallos y hojas)
plt.stem(x, p, basefmt=" ",linefmt='b-')
plt.title("Distribución de Pbabilidad Discreta\ncon Media y Desviación Estándar")
plt.xlabel("Valores de la variable aleatoria X")
plt.ylabel("P(X)")
plt.grid(True, linestyle='--', alpha=0.60)

#Dibujar línea vertical en la media
plt.axvline(x=esperanza, color='red', linestyle='--', label=f"Media (E[X])={esperanza:.2f}")
#Resaltar el rango de +-1 desviación estándar

plt.axvspan(esperanza - desviacion, esperanza + desviacion, color='red',alpha=0.2,
            label=f"+-1 Desv. Estándar (o= {desviacion:.2f}" )
#Mostrar leyenda y lámites

plt.legend()
plt.ylim(0,max(p)+0.05)
plt.xticks(x)

#Mostrar gráfica
plt.show()
```
