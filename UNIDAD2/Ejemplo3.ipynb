```python
#Combinaciones con repetición
import math
from itertools import combinations_with_replacement

dado=[1,2,3,4,5,6]
n =len(dado)
r=2 #¿Como agrupar?, de dos en dos por que hay dados
CnR= math.factorial(n+r-1) / (math.factorial(r) * math.factorial((n+r-1)-r))

print( "Existen ",CnR," posibles combinaciones con repetición del total de ",n,
      " caracteres ", " en  grupos de  ",r,"en",r)
print("Aquí las combinaciones con repetición de dos dados")
#Definir el conjunto de elementos
elementos = [1,2,3,4,5,6]

#Número de elementos a elegir en cada combinación
r=2

#Calcular las combinacionescon repetición
combinaciones_datos = list (combinations_with_replacement(dado, r))

#Imprimir las combinaciones
for combinacion in combinaciones_datos:
    print(combinacion)
```

```python
#Permutaciones con pyton
nombres=["a","b","c","d","e","f","g"]
n= len(nombres)
k_permutations = 2 # Define the number of elements to choose for permutation
num_permutations = mt.factorial(n) / (mt.factorial(n-k_permutations))
print("Existen ",num_permutations," posibles permutaciones sin repetición de ",n," nombres ", " en grupos de",k_permutations, "en",k_permutations)
print("Aqui las permutaciones")
permutaciones_nombres = list (permutations(nombres, k_permutations))
for permutacion in permutaciones_nombres:
    print(permutacion)
```
