```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leer el archivo
df = pd.read_csv('autos_corregidos.csv')
```

```python
#Mostrar las primeras filas
df.head()
```

```python
# Convertir price a numérico (por si hay errores en el dataset)
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Eliminar valores nulos
df = df.dropna(subset=['price'])
```

```python

df['price'].min()
```

```python
df['price'].max()
```

```python
#Calculamos la moda
df['price'].mode()
```

```python
#Calculamos la media
df['price'].mean()
```

```python
#Calculamos la mediana
df['price'].median()
```

```python
#Desviacion estándar
df['price'].std()
```

```python
#Estadisticos y Cuartiles con describe()
df['price'].describe()
```

```python
#Aplicamos la Regla de Sturges
n = len(df['price'])
k = int(1 + 3.322 * np.log10(n))
k
```

```python
#Realizamos el Histograma de precios
plt.figure(figsize=(8,5))
plt.hist(df['price'], bins=k)
plt.title('Histograma de Precios de Automóviles')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Realizamos un diagrama de Cajas y Bigotes. Este de relacion Precio-Vehiculo
plt.figure(figsize=(8,5))
sns.boxplot(x='body-style', y='price', data=df)

plt.title('Relación entre Precio y Tipo de Vehículo')
plt.xlabel('Tipo de Vehículo')
plt.ylabel('Precio')
plt.show()
```

```python
#Realizamos otro diagrama de Cajas y bigotes. Este de relacion Precio-Tipo de traccion
plt.figure(figsize=(8,5))
sns.boxplot(x='drive-wheels', y='price', data=df)

plt.title('Relación entre Precio y Tipo de Tracción')
plt.xlabel('Tipo de Tracción')
plt.ylabel('Precio')
plt.show()
```
