```python
import random

import matplotlib.pyplot as plt

from statistics import*
```

```python
fabrical = []

for i in range(100):

    numero_aleatorio = round(random.uniform(50, 100), 2)

    fabrical.append(numero_aleatorio)

print(fabrical)
```

```python
fabrica2 = []

for i in range(100):

    numero_aleatorio = round(random.uniform(48, 93), 2)

    fabrica2.append(numero_aleatorio)

print(fabrica2)
```

```python
plt.boxplot(fabrical)

plt.ylabel('Produccion en mts.')

plt.title('Produccion de la fabrica 1')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

bplot1 = ax1.boxplot(fabrical,
                     vert=True, # Alineacion Vertical
                     patch_artist=True) # Rellenar con color
ax1.set_title('Producción fabrica 1')

bplot2 = ax2.boxplot(fabrica2,
                     vert=True, # Alineacion Vertical
                     patch_artist=True) # Rellenar con color
ax2.set_title('Produccion fabrica 2')

plt.show()
```

```python
#Insertonos un ictar de 158 a lo fabrico 1.

#Seira un valor stipico

fabrical.append(150)

print(fabrical)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

bplot1 = ax1.boxplot(fabrical,
                     vert=True, # vertical box alignment
                     patch_artist=True) # fill with color
ax1.set_title('Producción fabrica 1')

bplot2 = ax2.boxplot(fabrica2,
                     vert=True, # vertical box alignment
                     patch_artist=True) # fill with color
ax2.set_title('Producción fabrica 2')

plt.show()
```

```python
# Agregamos ax1.set_ylim(30, 160) para garantizar que ambas gráficas usen la misma escala vertical, facilitando la comparación visual

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 5))

bplot1 = ax1.boxplot(fabrical,
                     vert=True, # Alineacion vertical
                     patch_artist=True) # Rellenar con color
ax1.set_title('Producción fabrica 1')
ax1.set_ylim(30, 160) # Escala uniforme de 30 a 160

bplot2 = ax2.boxplot(fabrica2,
                     vert=True, # Alineacion vertical
                     patch_artist=True) # Rellenar con color
ax2.set_title("Producción fabrica 2")
ax2.set_ylim(30, 160) # Escala uniforme de 30 a 160

plt.show()
```
